  Given the `root` of a binary tree, return all root to leaf paths in any order.

  ## Examples

  ### Example 1

  ![](./images/q1_ex1.png)

  Input: `root = [1, 2, 2, 3, 5, 6, 7]` *What traversal method is this?*

  Output: [[1, 2, 3], [1, 2, 5], [1, 2, 6], [1, 2, 7]]

  ### Example 2

  ![](./images/q1_ex3.png)

  Input: `root = [10, 9, 7, 8]`

  Output: [[10, 7], [10, 9, 8]]

</details>

#### Starter Code for Question 2

```python
# Definition for a binary tree node.
from typing import List

class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    if not root:
        return []
    result = []

    def dfs(node: TreeNode, path: List[int]) -> None:
        path.append(node.val)
        if node.left is None and node.right is None:
            result.append(path[:])
        else:
            if node.left:
                dfs(node.left, path)
            if node.right:
                dfs(node.right, path)
        path.pop()

    dfs(root, [])
    return result
```

```python
# Example: root = [1, 2, 2, 3, 5, 6, 7] (level-order: 1 has children 2,2; left 2 has 3,5,6,7)
# Build tree: root(1), left(2).left=3, left(2).right=2 with children 5,6,7
root = TreeNode(1, TreeNode(2, TreeNode(3), TreeNode(5)), TreeNode(2, TreeNode(6), TreeNode(7)))
print(bt_path(root))  # [[1, 2, 3], [1, 2, 5], [1, 2, 6], [1, 2, 7]]
```

```python
# New example (my own):
# Input: nums = [10, 20, 30, 20, 10]
# Scan: 10 (new), 20 (new), 30 (new), 20 (seen!) â†’ first duplicate is 20.
# Output: 20

# Trace of partner's example: nums = [4, 5, 6, 4, 6]
# We scan left to right. seen = {}.
# - 4: not in seen â†’ add 4, seen = {4}
# - 5: not in seen â†’ add 5, seen = {4, 5}
# - 6: not in seen â†’ add 6, seen = {4, 5, 6}
# - 4: 4 is in seen â†’ return 4 (first duplicate).
# So output is 4. The second 6 is never considered because we return as soon as we find the first duplicate.
```


-   Copy the solution your partner wrote. 


```python
# Partner's solution (set-based):
from typing import List

def set_first_duplicate(nums: List[int]) -> int:
    seen = set()
    for n in nums:
        if n in seen:
            return n
        seen.add(n)
    return -1
```


-   Explain why their solution works in your own words.


```python
# Why it works: We keep a set of values we have already seen. We walk the list once from left
# to right. The first time we see a value that is already in the set, that value is the
# first duplicate, so we return it immediately. If we finish the loop without returning,
# we never saw a duplicate, so we return -1. Using a set gives O(1) membership checks,
# and we only care about "seen or not," so we don't need counts.
```


-   Explain the problemâ€™s time and space complexity in your own words.


```python
# Time complexity: O(n) â€” we do a single pass over the list, and each set membership check
# and set add is O(1) on average.
# Space complexity: O(n) â€” in the worst case (no duplicate), the set holds all n elements.
# So both are linear in the input size, which is optimal for this problem since we must
# at least read the input and may need to store seen values until we find a duplicate.
```


-   Critique your partner's solution, including explanation, and if there is anything that should be adjusted.


```python
# Critique: The solution is correct and uses the right data structure (set) for O(1) lookups.
# One small adjustment: the partner's complexity explanation says "O(n) = O(1)" which is
# misleading â€” O(n) is linear, not constant. They likely meant "each operation is O(1)"
# for the per-element work. Otherwise the logic is clear and the set-based version is
# preferable to the list-based version (which would be O(n) time per "in" check, giving
# O(n^2) overall). No code changes are necessary; only the complexity wording could be
# clarified to avoid confusing O(n) with O(1).
```
